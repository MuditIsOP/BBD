```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

print("First 5 rows of the Iris dataset:")
display(iris_df.head())
```

---

```python
setosa_df = iris_df[iris_df['species'] == 'setosa']

plt.figure(figsize=(8, 6))
sns.regplot(
    data=setosa_df,
    x='sepal length (cm)',
    y='sepal width (cm)',
    scatter_kws={'alpha': 0.7, 'color': 'tab:blue'},
    line_kws={'color': 'red'}
)
plt.title('Linear Regression: Sepal Length vs Sepal Width (Setosa)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
versicolor_df = iris_df[iris_df['species'] == 'versicolor']

plt.figure(figsize=(8, 6))
sns.regplot(
    data=versicolor_df,
    x='sepal length (cm)',
    y='sepal width (cm)',
    scatter_kws={'alpha': 0.7, 'color': 'tab:green'},
    line_kws={'color': 'red'}
)
plt.title('Linear Regression: Sepal Length vs Sepal Width (Versicolor)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
virginica_df = iris_df[iris_df['species'] == 'virginica']

plt.figure(figsize=(8, 6))
sns.regplot(
    data=virginica_df,
    x='sepal length (cm)',
    y='sepal width (cm)',
    scatter_kws={'alpha': 0.7, 'color': 'tab:purple'},
    line_kws={'color': 'red'}
)
plt.title('Linear Regression: Sepal Length vs Sepal Width (Virginica)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
